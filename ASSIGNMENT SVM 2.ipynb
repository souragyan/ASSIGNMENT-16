```python
import pandas as pd
import numpy as np
#from sklearn.preprocessing import StandardScalar
from sklearn import svm
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.model_selection import train_test_split,cross_val_score
```

```python
df=pd.read_csv('salary.csv')
```

```python
df
```

```python
df.drop('educationno',axis='columns',inplace=True)
df.drop('maritalstatus',axis='columns',inplace=True)
df.drop('relationship',axis='columns',inplace=True)
df.drop('race',axis='columns',inplace=True)
df.drop('sex',axis='columns',inplace=True)
df.drop('capitalgain',axis='columns',inplace=True)
df.drop('capitalloss',axis='columns',inplace=True)
df.drop('native',axis='columns',inplace=True)
```

```python
df
```

```python
dummies=pd.get_dummies(df.workclass)
dummies.head(2)
df=pd.concat([df,dummies],axis='columns')
df.head(2)
```

```python
dummies=pd.get_dummies(df.education)
dummies.head(2)
df=pd.concat([df,dummies],axis='columns')
df.head(2)
```

```python
dummies=pd.get_dummies(df.occupation)
dummies.head(2)
df=pd.concat([df,dummies],axis='columns')
df.head(2)
```

```python
dummies=pd.get_dummies(df.Salary)
dummies.head(2)
```

```python
df=pd.concat([df,dummies],axis='columns')
df.head(2)
```

```python
#df.drop('workclass',axis='columns',inplace=True)
#df.drop('education',axis='columns',inplace=True)
#df.drop('occupation',axis='columns',inplace=True)
df.drop('Salary',axis='columns',inplace=True)
#df.drop('<=50K',axis='columns',inplace=True)
```

```python
df.shape
```

```python
x=df.iloc[:,0:39]
y=df.iloc[:,39]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=230)
```

```python
print(x_train.shape,y_train.shape)
print(x_test.shape,y_test.shape)
```

```python
#GRID SEARCH CV
clf=SVC()
param_grid=[{'kernel':['rbf'],'gamma':[50,10,0.5],'C':[10,0.15]}]
gsv=GridSearchCV(clf,param_grid,cv=10)
gsv.fit(x_train,y_train)
```

```python
gsv.best_params_,gsv.best_score_
```

```python
clf=SVC(C=10,gamma=50)
clf.fit(x_train,y_train)
y_pred=clf.predict(x_test)
```

```python
acc=accuracy_score(y_test,y_pred)*100
print("accuracy=",acc)
confusion_matrix(y_test,y_pred)
```
